### init packages

```python
%pip install matplotlib
%pip install pymongo

import numpy as np
import matplotlib.pyplot as plt

from pymongo import MongoClient
```

### init mongo

```python
client = MongoClient("localhost", 27017)

db = client.league_of_legends
```

### plot gold

```python
db_matches_gold = db.matches_gold
```

```python
gold_differences = np.array([0])
games = np.array([0])

for gold_difference in db_matches_gold.find():
    gold_differences = np.append(gold_differences, gold_difference['gold_difference'])
    games = np.append(games, games[-1] + 1)
```

```python
plt.style.use('dark_background')

fig, ax = plt.subplots()

ax.set_xlabel('game number')
ax.set_ylabel('gold adventage')
ax.set_title('league of legends gold adventage')
ax.plot(games, gold_differences, color='mediumseagreen', label='EMERALD I')
ax.legend()
plt.show()
```
